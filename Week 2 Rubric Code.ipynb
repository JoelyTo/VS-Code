```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Printing any Bracelet 

start = [1, 5] # can pick this as any ordered pair 
bracelet = []
s1 = start[0]
s2 = start[1]
bracelet.append(s1)
bracelet.append(s2)
max = 61 # used 61 as the upper bound as the longest sequence I found had a length of 60
for i in range(1, max): 
    next = (bracelet[i] + bracelet[i-1]) % 10
    bracelet.append(next)
```

```python
# Plotting any Bracelet 

x = [i for i in range(1, max)]
y = bracelet 
plt.plot(x, y[-max + 1:], '-ko', label = f'Bracelet B({s1, s2})')
plt.xlabel('Index number of the bracelet')
plt.ylabel('Bracelet value')
plt.grid('on')
plt.legend()
plt.show()


```

```python
# Counting how many different bracelets there are

n = 10
for j in range(0, n):
    m2 = j
    for i in range(0, n):
        m1 = i 
        bracelet = []
        bracelet.append(m1)
        bracelet.append(m2)
        bracelet_2 = []
        for a in range(0, n):
            n2 = a
            for b in range(0, n):
                n1 = b
                if (a != j) & (b != i):
                    bracelet_2.append(n1)
                    bracelet_2.append(n2)
        count = 0
        for p in range(0, max): 
            next = (bracelet[p] + bracelet[p-1]) % 10
            next_2 = (bracelet_2[p] + bracelet_2[p-1]) % 10
            bracelet.append(next)
            bracelet_2.append(next_2)
        for k in range(0, max):
            if bracelet[k] == bracelet_2[k]:
                count += 1
count
```

```python
# Extension: counting the number of different bracelets with a variety of mod n

n = 10
X = [i for i in range(2,200)]
c_a = []
for z in range(len(X)):
    x = X[z]
    for j in range(1, n):
        m2 = j
        for i in range(1, n):
            m1 = i 
            bracelet = []
            bracelet.append(m1)
            bracelet.append(m2)
            bracelet_2 = []
            for a in range(1, n):
                n2 = a
                for b in range(1, n):
                    n1 = b
                    if (a != j) & (b != i):
                        bracelet_2.append(n1)
                        bracelet_2.append(n2)
            count = 0
            for p in range(1, max): 
                    next = (bracelet[p] + bracelet[p-1]) % x
                    next_2 = (bracelet_2[p] + bracelet_2[p-1]) % x
                    bracelet.append(next)
                    bracelet_2.append(next_2)
            for k in range(1, max):
                if bracelet[k] == bracelet_2[k]:
                    count += 1
    c_a.append(count)

```

```python
# Extension part 2: Let's plot!

plt.scatter(X, c_a)
plt.xlabel('Mod')
plt.ylabel('Number of Different Bracelets')
plt.show()
```
