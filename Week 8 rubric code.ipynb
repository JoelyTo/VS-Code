```python
def PathNumber(target, iterations):
    x0 = 1
    paths = []
    x = x0

    for j in range(iterations):
        x = x0
        walk = []
        while(x < target+1):
            walk.append(x)
            rng = random.randint(0,1)
            if(rng == 0):
                x += 1
            else:
                x += 2
        if(walk[-1:][0] == target):
            paths.append(walk)
    return len(paths)

PathNumber(25, 20000)
```

```python
PathNumber(23, 10000) + PathNumber(24, 10000)
```

```python
nos = []
its = int(5e4)
for i in range(2,100):
    nos.append(PathNumber(i, its)/its)
plt.plot(nos)
plt.xlabel('Final Number, $X$')
plt.ylabel('Probability')
plt.title('Probability of landing on $X$')
```

```python
def f(N):
    pp = 0.5*(1 + np.sqrt(5))
    pm = 0.5*(1 - np.sqrt(5))
    A = (pm - 2)/(pp*pp*(pm - pp))
    B = (2 - pp)/(pm*pm*(pm - pp))
    return A*(pp**N) + B*(pm**N)
```

```python
def prob(n):
    A = 2 - 8/3
    return A*(-0.5)**n + 2/3
```

```python
x = [_ for _ in range(2, 30)]
vals = [2/3 - prob(_) for _ in range(2, 30)]
plt.figure(figsize  = (12,6))
plt.plot(x, vals)
plt.title('Difference of Probability and 2/3 - Convergence')
plt.show()
```

```python
its = 100

def ruinProb(n, p):
    total = 0
    while((n!=0)&(total < its)):
        c = random.randint(0,1000)
        if(c > p*1000):
            n -= 1
        else:
            n += 1
        total += 1
        print(n)
        
def ruinProbCount(n, p):
    total = 0
    while((n!=0)&(total < its)):
        c = random.randint(0,1000)
        if(c > p*1000):
            n -= 1
            total += 1
        else:
            n += 1
            total += 1
    return total
```

```python
ruinProbCount(10,0.1)
```

```python
cs = []
for i in range(1,20):
    a = []
    for j in range(100):
        a = np.append(a, ruinProbCount(i,0.5))
    cs = np.append(cs, np.mean(a))

plt.figure()
plt.plot(cs)
plt.xlabel('Starting Account')
plt.ylabel('Average Number of Moves until Ruin')
plt.show()
    
```

```python
plt.figure(figsize = (12,6))
for j in range(10):
    p = 0.1*j
    cs = []
    for i in range(1,20):
        a = []
        for j in range(100):
            a = np.append(a, ruinProbCount(i,p))
        cs = np.append(cs, np.mean(a))
    plt.plot(cs, label = f'$P =${round(p,1)}')
plt.xlabel('Starting Account')
plt.ylabel('Average Number of Moves until Ruin')
plt.title('Number of moves until ruin as both initial and probabilities vary')
plt.legend()
plt.show()
    
```
